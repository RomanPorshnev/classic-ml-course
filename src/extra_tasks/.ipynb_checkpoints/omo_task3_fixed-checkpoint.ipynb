```python
import numpy as np
from sklearn.metrics.pairwise import pairwise_distances

class DBSCAN:

    def __init__(self, data, eps, min_pts, metrics):
        self.data = data
        self.eps = eps
        self.min_pts = min_pts
        self.metrics = metrics
        self.labels = np.full(len(self.data), fill_value=-1, dtype=int)
        self.label = 0
        self.core_objects_not_checked, self.core_objects = [], []
        self.adjacency_matrix = pairwise_distances(X=self.data, Y=self.data,
                                               metric=self.metrics)


    def fit_transform(self):
        for curr_object in range(len(self.data)):
            neighbours = np.where(self.adjacency_matrix[
                curr_object] < self.eps)[0]

            if len(neighbours) >= self.min_pts:
                self.core_objects.append(curr_object)
                self.core_objects_not_checked.append(curr_object)

        if len(self.core_objects_not_checked) > 0:
            start_object = self.core_objects_not_checked[
                np.random.randint(0, len(self.core_objects_not_checked))]

            self.__search_for_connected_components(start_object)
        else:
            print("No core objects")

        return self.labels


    def __search_for_connected_components(self, curr_object):
        queue = [curr_object]
        self.labels[curr_object] = self.label
        while (len(queue) > 0):
            curr_object = queue.pop()
            neighbours = np.where(self.adjacency_matrix[
                curr_object] < self.eps)[0]

            if len(neighbours) >= self.min_pts:
                self.core_objects_not_checked.remove(curr_object)
                values_to_delete = [curr_object]
                for i, label in enumerate(self.labels):
                    if label != -1:
                        values_to_delete.append(i)
                mask = np.logical_not(np.isin(neighbours, values_to_delete))
                neighbours = neighbours[mask]
                for neighbour in neighbours:
                    self.labels[neighbour] = self.label
                queue.extend(neighbours.tolist())

        if len(self.core_objects_not_checked) > 0:
            start_object = self.core_objects_not_checked[
                np.random.randint(0, len(self.core_objects_not_checked))]
            self.label += 1
            self.__search_for_connected_components(start_object)
        else:
            return
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

n_objects = 500 # number of objects
data, blob_labels = make_blobs(n_samples=n_objects , centers=10, cluster_std=0.50, random_state=42)

dbscan = DBSCAN(data=data, eps=0.5, min_pts=10, metrics='euclidean')
labels = dbscan.fit_transform()
print(labels)

colors = ['black' if label == -1 else plt.cm.tab10(label) for label in labels] # шумы помечены чёрным цветом
plt.scatter(x=data[:, 0], y=data[:, 1], c=colors, s=10)
plt.title(label='Euclidean metric')
plt.show()
```

```python
dbscan = DBSCAN(data=data, eps=0.5, min_pts=10, metrics='manhattan')
labels = dbscan.fit_transform()
print(labels)

colors = ['black' if label == -1 else plt.cm.tab10(label) for label in labels] # шумы помечены чёрным цветом
plt.scatter(x=data[:, 0], y=data[:, 1], c=colors)
plt.title(label='Manhattan metric')
plt.show()
```
