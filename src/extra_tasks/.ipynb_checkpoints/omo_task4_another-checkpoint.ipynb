```python
import torch
from sklearn.cluster import KMeans

import seaborn as sns

import warnings

warnings.filterwarnings('ignore')
SEED = 42
```

# Один из алгоритмов кластеризации (07.04) - 3 бонуса
Реализовать для наблюдений, у которых все признаки вещественные числа, один из алгоритмов (Mean-shift, Affinity Propagation, Спектральная кластеризация). Подразумевается самостоятельное подробное изучение алгоритмов. В файле с решением должен быть пример кластеризации на данных с не плоской геометрией.


```python
class CoolSpectralClustering:
    def __init__(self, gamma = 1.0, seed = 42, n_clusters = 2, n_components = None):
        self.gamma = gamma
        self.n_clusters = n_clusters
        self.n_components = n_components

        self.seed = seed
        self.labels_ = None

    def fit(self, X):

        # Calculate Laplacian
        A = self._compute_similarity_matrix(X)
        assert A.shape == (len(X), len(X))

        L = self._compute_laplacian_matrix(A)

        # Get Normalized eigen vectors
        normalized_eigenvectors = self._compute_normalized_eigenvectors(L)

        # Fit K-means on eigen vectors
        kmeans = KMeans(n_clusters=self.n_clusters, random_state=self.seed)
        kmeans.fit(normalized_eigenvectors.detach().numpy())
        self.labels_ = kmeans.labels_


    def _compute_similarity_matrix(self, X):
        pairwise_distances = torch.cdist(X, X, p=2)

        # Affinity is RBF
        similarity_matrix = torch.exp(-self.gamma * pairwise_distances ** 2)
        return similarity_matrix

    def _compute_laplacian_matrix(self, similarity_matrix):
        degree_matrix = torch.diag(torch.sum(similarity_matrix, dim=1))

        laplacian_matrix = degree_matrix - similarity_matrix
        return laplacian_matrix

    def _compute_normalized_eigenvectors(self, laplacian_matrix):
        eigen_val, eigen_vec = torch.linalg.eigh(laplacian_matrix)

        # Compute norms of the eigenvalues and sort indices
        norms = torch.norm(eigen_val.squeeze().view(1, -1), dim=1)
        sorted_indices = torch.argsort(norms)

        # Select the top eigenvectors
        top_eigen_vec = eigen_vec[:, sorted_indices[:self.n_components]]

        return top_eigen_vec
```

```python
from sklearn.datasets import make_circles

X, y = make_circles(n_samples=[1000, 500],
                    noise=0.08,
                    factor=0.3,
                    random_state=SEED)

sns.scatterplot(x=X[:, 0], y=X[:, 1], c=y, cmap='Dark2')
```

```python
spectral = CoolSpectralClustering(gamma=6,
                                  n_components=None,
                                  n_clusters=2
                                )

spectral.fit(torch.tensor(X))

y_pred = spectral.labels_

sns.scatterplot(x=X[:, 0], y=X[:, 1], c=y_pred, cmap='Dark2')
```
